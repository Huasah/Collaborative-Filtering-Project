```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel("/Users/ryanmilligan/Downloads/[final] April 2015 to Nov 30 2019 - Transformed Jester Data - .xlsx", encoding = "ISO-8859-1")
```

```python
data.head()
```

```python
data = data.replace(99, np.NaN)
```

```python
sns.set_style('whitegrid')
```

```python
data.columns
```

```python
data.describe()
```

```python
sns.set_style('whitegrid')
sns.jointplot(x = '99.151', y = '99.152', data = data)
```

```python
plt.figure(figsize = (10,5))
plt.title('Distribution of ratings for joke 151')
sns.distplot(data['99.151'],bins=30,kde=False,color='red')
```

```python
new_jokes = data[['99.151', '99.152', '99.153', '99.154', '99.155', '99.156']]
```

```python
plt.figure(figsize = (10,5))
plt.title('Gas Plot')
sns.set_style('dark')
sns.violinplot(data = new_jokes)
sns.despine()
```

```python
sns.set_style('ticks')
plt.figure(figsize = (10,5))
sns.boxplot(data = new_jokes, palette = 'rainbow')
sns.despine()
```

```python
plt.figure(figsize = (8,5))
sns.heatmap(new_jokes.corr(), cmap = 'coolwarm')
plt.title('Correlation of Ratings Between New Jokes')
```

```python
sns.set_style('darkgrid')
sns.swarmplot(data = new_jokes)
```
